```python
dum_depart = pd.get_dummies(x['Department'])
```

```python
dum_depart
```

```python
x2 = x.copy()
x2.drop('Department',axis=1,inplace=True)
x3 = pd.concat([x2,dum_depart],axis=1)
```

```python
x3.head()
```

```python
positive.groupby('MaritalStatus')['Attrition'].count()
```

```python
labels_ordered_marr = positive.groupby('MaritalStatus')['Attrition'].count().sort_values(ascending=False).index
```

```python
labels_ordered_marr = {j:i for i,j in enumerate(labels_ordered_marr)}
x4 = x3.copy()
x4['MaritalStatus'] = x4['MaritalStatus'].map(labels_ordered_marr)
```

```python
x4.head()
```

```python
labels_time = positive.groupby('OverTime')['Attrition'].count().sort_values(ascending=False).index
labels_time = {j:i for i,j in enumerate(labels_time)}
x4['OverTime'] = x4['OverTime'].map(labels_time)
```

```python
categorical_features
```

```python
gender_dum = pd.get_dummies(x4['Gender'])
```

```python
x5 = x4.copy()
x5.drop('Gender',axis=1,inplace=True)
```

```python
x6 = pd.concat([x5,gender_dum],axis=1)
```

```python
x6.head()
```

```python
edu_dum = pd.get_dummies(x6['EducationField'])

```

```python
edu_dum.isnull().sum()
```

```python

x7 = pd.concat([x6,edu_dum],axis=1)
x7.drop('EducationField',axis=1,inplace=True)
x7.head()
```

```python
travel_label = positive.groupby('BusinessTravel')['Attrition'].count().sort_values(ascending=False).index
travel_label = {j:i for i,j in enumerate(travel_label)}
x7['BusinessTravel'] = x7['BusinessTravel'].map(travel_label)
```

```python
x7.head()
```

```python
x7.drop('BusinessTravel',axis=1,inplace=True)
```

```python
role_label = pd.get_dummies(x7['JobRole'])
```

```python
x7.drop('JobRole',axis=1,inplace=True)
```

```python
x8 = pd.concat([x7,role_label],axis=1)
```

```python
x8.head()
```

```python
x8.drop(['EmployeeNumber','DistanceFromHome'],axis=1,inplace=True)
```

```python
x8.drop('Id',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x8,y,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,max_depth=3,criterion='gini',random_state=1)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,rf.predict(x_test))
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=2500)
log.fit(x_train,y_train)
log.score(x_test,y_test)
```

```python
log.score(x_train,y_train)
```
